```python
%matplotlib inline
import neuron
import matplotlib.pyplot as plt
h = neuron.h
```

```python
soma = h.Section("soma")
soma.L = 15
soma.diam = 15
soma.nseg = 1

dend = h.Section("dend")
dend.L = 1000
dend.diam = 2
dend.nseg = int(dend.L/10)

dend.connect(soma,1,0)

for sec in h.allsec():
    sec.insert("hh")
    sec.Ra = 100
    sec.cm = 1
    
stim = h.IClamp(soma(0.5))
stim.delay = 50
stim.amp = 0.5
stim.dur = 100

t = h.Vector()
t.record(h._ref_t)
v = h.Vector()
v.record(soma(0.5)._ref_v)

h.dt = 0.01
h.finitialize(-65)
neuron.run(200)
plt.plot(t,v)
plt.show()
```
